```python
import numpy as np


# Calculate profit/loss from revenue and expenses

Revenue = np.array([[180,200,220],[24,36,40],[12,18,20]])
Expense = np.array([[80,90,100],[10,16,20],[8,10,10]])

PL = Revenue - Expense

print("Profit & Loss :\n", PL)
```

```python
# Calculate total sales from units and price per unit using matrix multiplication

Price = np.array([[1000,400,1200]])
Units = np.array([[30,40,50],[5,10,15],[2,5,7]])

TotalSales = Price.dot(Units)

print("Total Sales :\n", TotalSales)
```

```python
# Convert USD revenues to INR

USD = np.array([[200,220,250],[68,79,105],[110,140,180],[80,85,90]])
INR = USD * 75

print("INR :\n", INR)
```
