## MSE & MAE

```python
import numpy as np


# Guess this as a Predicted & True Value

y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])



# Mean Squared Error

SE = np.square(y_predicted - y_true)
MSE = SE.mean()
print("MSE :",MSE)


# Mean Absolute Error

MAE = np.abs(y_predicted - y_true)
MAE = MAE.mean()
print("\nMAE", MAE)


```

## Log Loss (or) Binary Cross Entropy

```python

def log(y_predicted, y_true):

    epsilon = 1e-15
    y_pred = [max(i,epsilon) for i in y_predicted]
    y_pred = [min(i,1-epsilon) for i in y_pred]
    y_pred = np.array(y_pred)

    return -np.mean(y_true*np.log(y_pred) + (1-y_true)*np.log(1-y_pred))


LogLoss = log(y_predicted, y_true)
print("Log Loss :", LogLoss)
```
