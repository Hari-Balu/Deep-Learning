```python
import math


# Sigmaoid

def sigmoid(x):
  return 1 / (1 + math.exp(-x))



print("Sigmoid 100 :", sigmoid(100))

print("Sigmoid 1 :", sigmoid(1))

print("Sigmoid -56 :", sigmoid(-56))

print("Sigmoid 0.5 :", sigmoid(0.5))
```

```python
# TanH

def tanh(x):
  return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))


print("TanH 100 :", tanh(100))

print("TanH -56 :", tanh(-56))

print("TanH 50 :", tanh(50))

print("TanH 1 :", tanh(1))
```

```python
# ReLU

def relu(x):
    return max(0,x)


print("ReLU -100 :", relu(-100))

print("ReLU 8 :", relu(8))
```

```python
# Leaky ReLU

def leaky_relu(x):
    return max(0.1*x,x)


print("Leaky ReLU -100 :", leaky_relu(-100))

print("Leaky ReLU 8 :", leaky_relu(8))
```
